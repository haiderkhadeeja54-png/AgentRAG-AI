```python
# ==========================================================
# 4. Load PDF
# ==========================================================
from google.colab import files

uploaded = files.upload()
```

```python
from langchain_community.document_loaders import PyPDFLoader

pdf_path = list(uploaded.keys())[0]

loader = PyPDFLoader(pdf_path)
documents = loader.load()

print(f"Number of pages loaded: {len(documents)}")
```

```python
print(documents[0].page_content[:1000])
```

```python
# ==========================================================
# 5. Split Documents
# ==========================================================
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = text_splitter.split_documents(documents)

print(f"Total chunks created: {len(chunks)}")
```

```python
# ==========================================================
# 6. Create Embeddings
# ==========================================================
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("✅ Embedding model loaded successfully!")
```

```python
# ==========================================================
# 7. Build Vector Database (FAISS)
# ==========================================================
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(
    documents=chunks,
    embedding=embeddings
)

print("✅ FAISS Vector Database Created!")
```

```python
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)

print("✅ Retriever Ready!")
```

```python
# ==========================================================
# 9. PDF Question Answering
# ==========================================================
query = "What is the objective of this project?"

results = retriever.invoke(query)

for i, doc in enumerate(results):
    print("="*80)
    print(f"Result {i+1}")
    print(f"Page: {doc.metadata['page']}")
    print(doc.page_content[:600])
    print()
```

```python
from typing import TypedDict

class AgentState(TypedDict):
    question: str
    answer: str
    source: str
```

```python
def pdf_tool(question):

    docs = retriever.invoke(question)

    context = "\n\n".join(
        doc.page_content for doc in docs
    )

    answer = chain.invoke({
        "context": context,
        "question": question
    })

    pages = sorted(
        set(doc.metadata["page"] + 1 for doc in docs)
    )

    return {
        "answer": answer,
        "source": f"PDF (Pages: {', '.join(map(str, pages))})"
    }
```

```python
pdf_tool("What GAN architecture is used?")
```

```python
def web_tool(question):

    results = web_search.invoke(question)

    context = "\n\n".join(
        r["content"] for r in results
    )

    answer = llm.invoke(
        f"""
Answer the following question using ONLY the web search results.

Question:
{question}

Search Results:
{context}
"""
    )

    return {
        "answer": answer.content,
        "source": "Live Web Search (Tavily)"
    }
```

```python
web_tool("Who is the CEO of Google?")
```

```python
def router(question):

    docs = retriever.invoke(question)

    context = "\n\n".join(
        doc.page_content for doc in docs
    )

    decision_prompt = f"""
You are an intelligent routing agent.

Determine whether the uploaded PDF contains enough information
to answer the user's question.

If YES reply with only:

PDF

If NO reply with only:

WEB

Question:
{question}

Retrieved Context:
{context}
"""

    decision = llm.invoke(decision_prompt).content.strip().upper()

    return decision
```

```python
router("What GAN architecture is used?")
```

```python
router("Who is the CEO of Google?")
```

```python
def agent(question):

    decision = router(question)

    print(f"🧠 Agent Decision: {decision}")

    if decision == "PDF":
        return pdf_tool(question)

    else:
        return web_tool(question)
```

```python
agent("What GAN architecture is used?")
```

```python
agent("Who is the CEO of Google?")
```

```python
from langgraph.graph import StateGraph, END

print("✅ LangGraph Imported Successfully")
```

```python
from typing import TypedDict

class AgentState(TypedDict):
    question: str
    decision: str
    answer: str
    source: str
```

```python
def router_node(state):

    question = state["question"]

    decision = router(question)

    return {
        "question": question,
        "decision": decision
    }
```

```python
def pdf_node(state):

    result = pdf_tool(state["question"])

    return {
        "question": state["question"],
        "decision": "PDF",
        "answer": result["answer"],
        "source": result["source"]
    }
```

```python
def web_node(state):

    result = web_tool(state["question"])

    return {
        "question": state["question"],
        "decision": "WEB",
        "answer": result["answer"],
        "source": result["source"]
    }
```

```python
router_node({
    "question":"What GAN architecture is used?"
})
```

```python
pdf_node({
    "question":"What GAN architecture is used?"
})
```

```python
web_node({
    "question":"Who is the CEO of Google?"
})
```

```python
# ==========================================================
# 11. LangGraph Agent
# ==========================================================
workflow = StateGraph(AgentState)

workflow.add_node("router", router_node)
workflow.add_node("pdf", pdf_node)
workflow.add_node("web", web_node)
```

```python
workflow.set_entry_point("router")
```

```python
workflow.add_conditional_edges(
    "router",
    lambda state: state["decision"],
    {
        "PDF": "pdf",
        "WEB": "web"
    }
)
```

```python
workflow.add_edge("pdf", END)
workflow.add_edge("web", END)
```

```python
graph = workflow.compile()

print("✅ LangGraph Compiled Successfully")
```

```python
graph.invoke({
    "question": "What GAN architecture is used?"
})
```

```python
graph.invoke({
    "question": "Who is the CEO of Google?"
})
```

```python
def ask_agent(question):

    result = graph.invoke({
        "question": question
    })

    print("=" * 70)
    print("🤖 AgentRAG AI")
    print("=" * 70)

    print(f"\n❓ Question:\n{question}")

    print(f"\n🧠 Decision:\n{result['decision']}")

    print(f"\n✅ Answer:\n{result['answer']}")

    print(f"\n📚 Source:\n{result['source']}")
```

```python
ask_agent("What GAN architecture is used?")
```

```python
ask_agent("Who is the CEO of Google?")
```

```python
def agent_response(question):
    """
    Returns the complete response from AgentRAG AI.
    """

    result = graph.invoke({
        "question": question
    })

    return {
        "question": question,
        "decision": result["decision"],
        "answer": result["answer"],
        "source": result["source"]
    }
```

```python
response = agent_response(
    "What GAN architecture is used?"
)

response
```
